```python
import numpy as np
import pandas as pd
from pandas import Series
from pandas import DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as stats
import scipy.stats as zscore

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('fifa.csv',sep=',')
```

```python
df.head()
```

```python
df.info()
```

```python
ser1 = Series(df.columns)
```

```python
df.drop(['Photo','Flag','Club Logo','Jersey Number','Preferred Foot','Skill Moves','Weak Foot'],axis=1,inplace=True)
df.drop(ser1[25:59],axis=1,inplace=True)
```

```python
df.describe()
```

## dropping null values 

```python
df.dropna(how='all',inplace=True)
```

```python
for i in range(len(df.columns)):
       if df.iloc[:,i].isnull().sum() <= 251:
            df.dropna(subset=[df.columns[i]],inplace=True)
        
```

```python
df.isnull().sum()
```

```python
df.reset_index(inplace=True)
```

```python
df.drop(['index'],axis=1,inplace=True)
```

## Converting value to float from object

```python
df['Value'] = df['Value'].str.lstrip('€')
```

```python
lst = []
for i in df['Value'].values:
    if i.endswith('M'):
        i = float(i.rstrip('M'))*(10**6)
        lst.append(i)
    elif i.endswith('K'):
        i = float(i.rstrip('K'))*(10**3)
        lst.append(i)
    else:
        lst.append(i)
```

```python
df['Value'] = lst
```

```python
df['Value'] = df['Value'].astype(float)
```

## Converting wage to float from object

```python
df['Wage'] = df['Wage'].str.lstrip('€').str.rstrip('K')
df['Wage'] = df['Wage'].astype(float)
df['Wage']*= 1000
```

## Converting Joined to float from object

```python
v = {'Joined':0}
df.fillna(value=v,inplace=True)
```

```python
lst = []
for i in df['Joined'].values:
    if i==0:
        lst.append(i)
    elif ',' in i:
        b = i.split(',')
        i = b[1].lstrip()
        lst.append(i)
```

```python
df['Joined'] = lst
```

```python
df['Joined'] = df['Joined'].astype(int)
```

## Converting Height to float from object

```python
lst = []
var = []
for i in df['Height'].values:
    var = i.split('\'')
    lst.append(float(var[0])*30.0+float(var[1])*2.54)
```

```python
df['Height'] = lst
```

## Converting weight to float from object

```python
df['Weight'] = df['Weight'].str.replace('lbs','').astype(float)
```

## Converting Release Clause to float from object

```python
df['Release Clause'] = df['Release Clause'].str.lstrip('€')
```

```python
v = {'Release Clause':0}
df.fillna(value=v,inplace=True)
```

```python
lst = []
for i in df['Release Clause'].values:
    if i==0:
        lst.append(i)
    elif i.endswith('M'):
        i = float(i.rstrip('M'))*(10**6)
        lst.append(i)
    elif i.endswith('K'):
        i = float(i.rstrip('K'))*(10**3)
        lst.append(i)
    else:
        lst.append(i)
```

```python
df['Release Clause'] = lst
```

```python
df['Release Clause'] = df['Release Clause'].astype(float)
```

## Converting Contract valid until to datetime from object

```python
df['Contract Valid Until'] = pd.to_datetime(df['Contract Valid Until'])
```
